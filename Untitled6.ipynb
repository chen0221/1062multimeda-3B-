```python
from skimage import io,data
import matplotlib.pyplot as plt
import numpy as np

img=data.chelsea()
img2=img.copy()

plt.figure(num='chelsea',figsize=(8,8)) #1 inch = 0.39cm
plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('modified image')

for i in range(0,50):
 for j in range(0,50):
   img2[i,j,1] = 0

plt.imshow(img2)
plt.show()
```

```python
from skimage import io,data
import matplotlib.pyplot as plt

img=data.chelsea()
rows,cols,dims=img.shape

img2 = img.copy()

plt.figure(num='chelsea',figsize=(8,8))
plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(img)

for i in range(0,rows):
 for j in range(0,cols):
   img2[i,j,2] = 0.5*img2[i,j,2]

plt.subplot(1,2,2)
plt.title('modified image')
plt.imshow(img2)
plt.show()
```

```python
from skimage import io,data

img=io.imread('black cat.jpg')
rows,cols,dims=img.shape
img2 = img[:,:,0]

for i in range(0,rows):
 for j in range(0,cols):
   sum = 0.0
   for k in range(0,3): #dims=3
    sum = sum + img[i,j,k]
   img2[i,j] = sum/3 #dims=3
io.imshow(img2)
io.show()

io.imsave('black cat2.jpg',img2)
```
